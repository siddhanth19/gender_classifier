```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import tensorflow as tf
from tensorflow import keras
```

## Importing Data

## reading the data

```python
DATADIR="C:/Users/Siddhanth/PycharmProjects/gender"
CATEGORIES=["MALE","FEMALE"]
for category in CATEGORIES:
    path=os.path.join(DATADIR,category)   #path to male or female
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap="gray")
        plt.show()
        break
    break
       
```

## resizing

```python
IMG_SIZE=100
new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array,cmap="gray")
```

## creating training data

```python
training_data=[]
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR,category)   #path to male or female
        class_num=CATEGORIES.index(category)  #assigning them a numerical value
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
create_training_data()
```

```python
print(len(training_data))
```

## shuffling the data

```python
import random
random.shuffle(training_data)
```

```python
for sample in training_data[:10]:
    print(sample[1])
```

```python
x=[]
y=[]
```

## FEATURE LABEL SPLITTING

```python
for features,label in training_data:
    x.append(features)
    y.append(label)
x=np.array(x).reshape(-1,IMG_SIZE,IMG_SIZE,1)
```

```python
plt.figure()
plt.imshow(x[0])
plt.colorbar()
plt.show()
```

## scaling it from 0 to 1 instead of 0 to 200

```python
x=x/225.0
x
```

```python
plt.figure()
plt.imshow(x[0])
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x[i],cmap="gray")
    plt.xlabel(y[i])
plt.show()
    
```

## building our neural network

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(IMG_SIZE,IMG_SIZE)),
    keras.layers.Dense(128,activation=tf.nn.relu),
    keras.layers.Dense(128,activation=tf.nn.relu),
    keras.layers.Dense(2,activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
y=np.array(y)
model.fit(x,y,epochs=10)
```

## extracting testing data

```python
testing_data=[]
CATEGORIES=['MALE','FEMALE']
DATADIR2="C:/Users/Siddhanth/PycharmProjects/gender/Validation"
def create_testing_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR2,category)   #path to male or female
        class_num=CATEGORIES.index(category)  #assigning them a numerical value
        for img in os.listdir(path):
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                testing_data.append([new_array,class_num])
create_testing_data()
import random
random.shuffle(testing_data)
for i in testing_data[:10]:
    print(i[1])
```

```python
x_test=[]
y_test=[]
for features,labels in testing_data:
    x_test.append(features)
    y_test.append(labels)
x_test=np.array(x_test).reshape(-1,IMG_SIZE,IMG_SIZE,1)
y_test=np.array(y_test)
```

```python
test_loss,test_acc=model.evaluate(x_test,y_test)
```

```python
def verify(i,y_test,x_test,pred):
    y_test,x_test,pred=y_test[i],x_test[i],pred[i]
    plt.xticks=([])
    plt.yticks=([])
    plt.figure(figsize=(5,3))
    plt.grid(False)
    plt.imshow(x_test,cmap="gray")
    predicted_label=np.argmax(pred)
    if(predicted_label==y_test):
        color="blue"
    else:
        color="red"
    plt.xlabel("{} {:2.0f}%({})".format(CATEGORIES[predicted_label],100*np.max(pred),CATEGORIES[y_test]),color=color)
    plt.show()
```

```python
num_img=10
for i in range(num_img):
    plt.xticks=([])
    plt.yticks=([])
    verify(i,y_test,x_test,pred)
plt.show()
```
